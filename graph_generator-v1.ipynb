```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Radial Graph Plotter

This notebook demonstrates how to create a radial (polar) graph using matplotlib. You can input a table with headings (angles) and distances from the center, and visualize the data as a polar plot.

```python
# Sample data: headings (degrees) and distances

data = {
    'Heading': [0, 45, 90, 135, 180, 225, 270, 315],
    'Distance': [5, 7, 6, 8, 5, 7, 6.8, 8]
}
df = pd.DataFrame(data)
df
```

```python
# Parameters for customizing the plot
heading_labels = [0, 90, 180, 270]  # Headings to mark (degrees)
ring_step = 2                       # Distance between rings
max_distance = 10                   # Maximum distance displayed
unit = 'metres'                     # Distance unit label
point_color = "#ffffff"            # Colour for points
ring_color = "#5dfd00"              # Colour for rings
label_color = "#ffffff"               # Colour for labels
bg_color = "#00000073"                # Background colour

# Plot radial (polar) graph with custom parameters
headings_rad = np.deg2rad(df['Heading'])
distances = df['Distance']

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
fig.patch.set_facecolor(bg_color)
ax.set_facecolor(bg_color)
ax.scatter(headings_rad, distances, marker='o', color=point_color)
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_title(f'Radar chart', color=label_color)

# Set custom heading labels
ax.set_xticks(np.deg2rad(heading_labels))
ax.set_xticklabels([f'{h}' for h in heading_labels], color=label_color)

# Set custom rings
ax.set_yticks(np.arange(0, max_distance + ring_step, ring_step))
ax.set_ylim(0, max_distance)
ax.set_ylabel("") #(f'Distance ({unit})', labelpad=20, color=label_color)
for spine in ax.spines.values():
    spine.set_color(ring_color)
    spine.set_linewidth(1.5)
ax.tick_params(axis='y', colors=label_color)

plt.show()
```
